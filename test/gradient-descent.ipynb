# Gradient Descent algorithm 

### Example y = x^2 + 1

```python
import matplotlib.pyplot as plt
import random
```

```python
x_init = random.randrange(100)
alpha = 0.25
n_iter = 100
```

```python
y = []
iterations = []
```

```python
x = x_init
for i in range(n_iter):
    # print("-" * 10)
    # print(f"iter #{i + 1}")
    
    gradient = 2 * x
    x = x - alpha * gradient
    
    y.append(x ** 2 + 1)
    iterations.append(i + 1)

    # print(f"x = {x}, y = {y[i]}")
```

```python
plt.plot(iterations, y)
plt.xlabel("Iteraciones")
plt.ylabel("y")
```
