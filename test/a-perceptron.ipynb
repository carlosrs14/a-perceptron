# A perceptron

```python
import numpy as np
```

## Perceptron class

```python
class Perceptron:
    def __init__(self, learning_rate = 0.1, epochs = 5):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = 0.0
    
    def activation(self, x):
        return 1 if x >= 0 else 0
    
    def fit(self, X: np.matrix, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        for _ in range(self.epochs):
            for i in range(n_samples):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_predicted = self.activation(linear_output)
                
                error = y[i] - y_predicted
                self.weights += self.lr * error * X[i]
                self.bias += self.lr * error

    def predict(self, X : np.matrix):
        y_predicteds = []
        for x in X:
            linear_output = np.dot(x, self.weights) + self.bias
            y_predicteds.append(self.activation(linear_output))
        return np.array(y_predicteds)
    

```

## Usage Exmaple

### Fit

```python
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 0, 0, 1])

perceptron = Perceptron(learning_rate=0.01, epochs=10)

perceptron.fit(X, y)
```

### Predict

```python
predictions = perceptron.predict(X)
print("Input data:", X, sep="\n")


# print("\nWeights:", perceptron.weights)
# print("Bias:", perceptron.bias)

print("\nExpected:", y)
print("Predictions:", predictions)

asserts = (y == predictions)
total = y.shape[0]
correct = np.sum(asserts)
print("Acertó en :", correct, "de ", total)
```
